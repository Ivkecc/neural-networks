# Neural Networks – Proces razvoja projekta



# 1. Problem

Zadatak projekta bio je klasifikacija aktivnosti čoveka (HAR) korišćenjem više arhitektura neuronskih mreža.




# Skup podataka

Za realizaciju projekta korišćen je **UCI Human Activity Recognition (HAR) Dataset**.

Skup podataka sadrži merenja prikupljena pomoću akcelerometra i žiroskopa ugrađenih u pametni telefon. Na osnovu tih merenja cilj je klasifikovati aktivnost korisnika.

U projektu se klasifikuju sledeće aktivnosti:

- Walking
- Walking Upstairs
- Walking Downstairs
- Sitting
- Standing
- Laying

Podaci se učitavaju kroz `Dataset` i `DataLoader`, nakon čega se koriste za treniranje i evaluaciju različitih neuronskih mreža.

# 2. Plan razvoja

Redosled implementacije:

1. DenseNN
2. CNN
3. LSTM
4. GRU
5. CNN + LSTM
6. Poređenje modela


# Arhitektura projekta

Tok izvršavanja projekta može se prikazati na sledeći način:

```text
Dataset
    │
    ▼
DataLoader
    │
    ▼
Selected Model
(DenseNN / CNN / LSTM / GRU / CNN+LSTM)
    │
    ▼
Training
    │
    ▼
Evaluation
    │
    ▼
Plots & Reports


# Hiperparametri i podešavanja

U svim eksperimentima korišćena je ista osnovna infrastruktura za treniranje modela.

| Parametar | Vrednost |
|------------|-----------|
| Optimizer | Adam |
| Loss funkcija | CrossEntropyLoss |
| Early Stopping | patience = 10 |
| Evaluacija | Accuracy, Confusion Matrix, Classification Report |
| Learning rate | 0.0005 |
| Weight decay | 1e-4 |


Korišćenje iste trening infrastrukture omogućava fer poređenje različitih arhitektura neuronskih mreža.

# 3. DenseNN

## Cilj

DenseNN je implementiran kao bazni model za poređenje.

### Eksperiment

- Implementacija trening petlje
- Early stopping
- Checkpointing
- Vizualizacija rezultata

### Rezultat
<div style="display:flex; justify-content:center; gap:8px; align-items:center;">
    <img src="outputs/plots/dense_accuracy.png" height="300">
    <img src="outputs/plots/dense_loss.png" height="300">
    <img src="outputs/plots/dense_confusion_matrix.png" height="300">
</div>

### Zaključak

Dense Neural Network predstavlja dobru početnu arhitekturu i ostvaruje približno 84–85 tačnosti na test skupu. Rezultati pokazuju da model uspešno prepoznaje dinamičke aktivnosti, dok najveći broj grešaka nastaje kod statičkih aktivnosti (Sitting, Standing i Laying). 


# 4. CNN

## Razlog implementacije

CNN je dodat kako bi bolje prepoznao lokalne obrasce u vremenskim signalima.

## Arhitektura

```text
Conv1d
↓
BatchNorm
↓
ReLU
↓
MaxPool
↓
Conv1d
↓
BatchNorm
↓
ReLU
↓
MaxPool
↓
Flatten
↓
Linear
↓
Output
```

# 9. Zaključak

Na osnovu sprovedenih eksperimenata može se zaključiti da je CNN model ostvario najbolje rezultate, ostvarivši najveću tačnost klasifikacije i najstabilnije ponašanje tokom treninga. DenseNN predstavlja dobar bazni model, dok su rekurentne arhitekture pokazale potencijal za obradu sekvencijalnih podataka, ali u ovoj implementaciji nisu nadmašile CNN. Budući razvoj projekta može biti usmeren na optimizaciju hiperparametara, primenu naprednijih arhitektura i razvoj sistema za klasifikaciju aktivnosti u realnom vremenu.

# Naučene lekcije

Tokom razvoja projekta pokazalo se da izbor arhitekture ima veliki uticaj na performanse modela.

DenseNN je poslužio kao dobar bazni model, dok je CNN ostvario najbolje rezultate zahvaljujući sposobnosti izdvajanja lokalnih obrazaca iz senzorskih podataka. Rekurentni modeli, LSTM i GRU, nisu ostvarili očekivane rezultate u ovoj implementaciji, dok je CNN+LSTM pokazao stabilne performanse, ali nije nadmašio samostalni CNN.

Takođe se pokazalo da složenija arhitektura ne mora nužno ostvariti bolje rezultate ukoliko hiperparametri nisu odgovarajuće podešeni. Zbog toga je evaluacija kroz accuracy, confusion matrix i classification report bila važna za razumevanje ponašanja svakog modela.
